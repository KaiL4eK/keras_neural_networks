```python
## Interesting articles:
# https://elitedatascience.com/dimensionality-reduction-algorithms
# http://sebastianraschka.com/Articles/2014_python_lda.html#principal-component-analysis-vs-linear-discriminant-analysis

import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import preprocessing, covariance

data = np.load('data.npz')

X = data['data']
y = data['classes']

print('Data obtain done')

std_scale = preprocessing.StandardScaler().fit(X.astype(np.float64))
X = std_scale.transform(X.astype(np.float64))
# print('Mean:', std_scale.mean_)
# print('Scale:', std_scale.scale_)

print('Data preparation done')

cov_est = covariance.EmpiricalCovariance().fit(X)
variances = cov_est.covariance_.diagonal()

print('Classes:', np.unique(y))
print('Data shape:', X.shape)

n_components = 200

if 0:
    pca3 = PCA(n_components=n_components).fit(X)
    X = pca3.transform(X)
    
    print('Covariance:\n', cov_est.fit(X).covariance_)

    print('Explained variance:\n', pca3.explained_variance_ratio_)
    print(np.sum(pca3.explained_variance_ratio_))
else:
    clf = LinearDiscriminantAnalysis(n_components=n_components).fit(X, y)
    X = clf.transform(X)
    
    print('Explained variance:\n', clf.explained_variance_ratio_)
    print('Explained variance len:\n', len(clf.explained_variance_ratio_))    
    print(np.sum(clf.explained_variance_ratio_))
    
if n_components == 2:

    print('Draw plot')
    class_data = []
    data = []

    for u_cls in np.unique(y):
        class_data.append(X[y == u_cls])

    for i, cls in enumerate(class_data):
        trace = go.Scatter(
            x=cls[:, 0],
            y=cls[:, 1],
            mode='markers',
            marker=dict(
                color=i,
                size=4,
                symbol='circle',
            )
        )
        data.append(trace)

    py.iplot(data, filename='data-scatter-3d')
```

```python
# Images data preparation

import os
import cv2
import numpy as np
import csv

img_size = (60, 60)

print('Start')

csv_file_paths = (os.path.join(root, filename)
                    for root, _, filenames in os.walk('GTSRB_data')
                    for filename in filenames if filename.endswith('.csv'))

imgs = np.ndarray(shape=(0, img_size[0]*img_size[1]), dtype=np.uint8)
ids = np.ndarray(shape=(0), dtype=np.uint8)

for file in csv_file_paths:
    print('Processing: %s' % file)
    with open(file, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
        next(spamreader)
        for row in spamreader:
            filepath = os.path.join(os.path.dirname(file), row[0])
            cls = int(row[7])
            
            img = cv2.imread(filepath)
            img = cv2.resize(img, img_size)
            cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            fl_cv_rgb = cv_rgb.flatten()

            imgs = np.append(imgs, [fl_cv_rgb], axis=0)
            ids  = np.append(ids, [cls], axis=0)

idx_s   = np.argsort(ids)
     
imgs = np.array(imgs)[idx_s]
ids = np.array(ids)[idx_s]

np.savez('data.npz', data=imgs, classes=ids)

print('Classes:', np.unique(ids))
print(imgs.shape, ids.shape)
print('Ready')
```
